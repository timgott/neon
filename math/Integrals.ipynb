## Code Generation

```python
import sympy
```

```python
from sympy.codegen.ast import CodeBlock, Assignment, Return
from sympy.codegen.rewriting import create_expand_pow_optimization, ReplaceOptim
```

```python
def myreplace(expr, filter, f):
    # normal replace will ignore to convert 2->2.0
    keys = expr.find(filter)
    replacemap = { k: f(k) for k in keys }
    return expr.xreplace(replacemap)

inversesqrt = sympy.Function("inversesqrt")

def sqrt_rewrite(expr):
    # for sympy, sqrt(x) == pow(x,0.5)
    # but sqrt should be faster than pow in glsl
    mysqrt = sympy.Function("sqrt")
    expr = myreplace(expr,lambda p: p.is_Pow and p.exp == -1/2, lambda p: sympy.Integer(1)/mysqrt(p.base))
    return expr

def int_float_rewrite(expr):
    # glsl does not convert int -> float implicitly
    return myreplace(expr, lambda p: p.is_Integer or p.is_Rational, lambda p: sympy.Float(p))

def compact_factors(expr):
    # try to rewrite e.g. x*x + x -> x*(x+1)
    def hornerreplace(p):
        try:
            before = p.count_ops()
            new_p = sympy.horner(p)
            after = new_p.count_ops()
            return new_p if after < before else p
        except sympy.PolynomialError:
            print("compact_factors failed")
            return None
    res, replacemap = expr.replace(lambda p: p.is_Add, hornerreplace, map=True)
    return expr.xreplace(replacemap)

def generate_glsl_codeblock(expr, compact=True, cse=True, expand_pow=True):
    expand_pow_opt = create_expand_pow_optimization(5)
    codeblock = CodeBlock(Assignment(sympy.symbols("result"), expr._sympy_()))
    if compact:
        codeblock = compact_factors(codeblock.simplify())
    if cse:
        codeblock = codeblock.cse(optimizations="basic") # reduce common subexpressions
    if compact:
        codeblock = compact_factors(codeblock)
    codeblock = sqrt_rewrite(codeblock)
    if expand_pow:
        codeblock = expand_pow_opt(codeblock)
    codeblock = int_float_rewrite(codeblock)
    return codeblock

def generate_glsl(expr, **kwargs):
    codeblock = generate_glsl_codeblock(expr, **kwargs)
    return sympy.ccode(codeblock, human=True)
```

```python
print(generate_glsl(arc_light_F))
```

```python
print(generate_glsl(neon_line_h))
```

```python
print(generate_glsl(neon_line))
```

```python
print(generate_glsl(neon_line.simplify(algorithm="fricas").simplify(), compact=False))
```
