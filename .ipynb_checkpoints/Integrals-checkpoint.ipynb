```python
from sympy.codegen.ast import CodeBlock, Assignment, Return
from sympy.codegen.rewriting import create_expand_pow_optimization, ReplaceOptim
```

```python
sympy.Float(sympy.Integer(2))
```

```python
int_to_float_opt = ReplaceOptim(lambda p: p.is_Integer, lambda p: sympy.Float(p))
```

```python
int_to_float_opt(arc_light_F._sympy_())
```

```python
int_test_opt = ReplaceOptim(lambda p: p.is_Integer, lambda p: print(type(p)))
```

```python
int_test_opt(arc_light_F._sympy_())
```

```python
int_test_opt(int_to_float_opt(arc_light_F._sympy_()))
```

```python
arc_light_F._sympy_()
```

```python
print(sympy.ccode((arc_light_F._sympy_()), type_aliases={int: sympy.codegen.ast.float32}))
```

```python
def generate_glsl(expr):
    expand_pow = create_expand_pow_optimization(5)
    codeblock = CodeBlock(Assignment(sympy.symbols("result"), expand_pow(expr._sympy_())))
    return sympy.ccode(expand_pow(codeblock.cse(optimizations="basic")), human=True)
```

```python
print(generate_glsl(arc_light_F))
```

```python
print(generate_glsl(neon_line))
```

```python
print(generate_glsl(point_light))
```
